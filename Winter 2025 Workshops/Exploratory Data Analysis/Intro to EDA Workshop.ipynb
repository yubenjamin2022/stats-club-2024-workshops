#### Identifying Outliers

Outliers are data which are far from the rest of the data. The general accepted way to find outliers are if the are 3x away from the 1st or 3rd quartiles, and is a suspected outlier if it is 1.5x away from the 1st or 3rd quartiles. That is:

$$x \geq 3 \times IQR + Q_3$$
$$x \leq Q_1 - 3 \times IQR$$

for definite outliers and:

$$x \geq 1.5 \times IQR + Q_3$$
$$x \leq Q_1 - 1.5 \times IQR$$

for suspected outliers, where:

$$IQR = Q_3 - Q_1$$

```python
Q1 = np.percentile(student_data['Hours_Studied'], 25)
Q3 = np.percentile(student_data['Hours_Studied'], 75)
IQR = Q3 - Q1
print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")

outliers = student_data['Hours_Studied'][
    (student_data['Hours_Studied'] >= Q3 + 3 * IQR) |
    (student_data['Hours_Studied'] <= Q1 - 3 * IQR)
]

sus_outliers = student_data['Hours_Studied'][
    (student_data['Hours_Studied'] >= Q3 + 1.5 * IQR) |
    (student_data['Hours_Studied'] <= Q1 - 1.5 * IQR)
]
```

##### Linear Regression

Linear regression is a common tool used to predict certain variables (the dependant variable), by making the assumption that the dependant variable is linearly related to the independent variables.

In layman's terms, we use our data to predict: $$\beta_0, \dots, \beta_n$$

which is used in the following equation:

$$y = \beta_0 + \beta_1\times x_1 + \dots + \beta_n\times x_n$$

```python
from sklearn.linear_model import LinearRegression
```

```python
X = student_data[['Hours_Studied', 'Sleep_Hours', 'Attendance']]  # Independent Variables
y = student_data['Exam_Score']  # Dependent Variable
```

```python
model = LinearRegression()  # Initialize the model
model.fit(X, y)  # Train the model
```

```python
print("Intercept:", model.intercept_)  # Bias term (b0)
print("Coefficients:", model.coef_)   # Slopes for each feature (b1, b2, b3)
```

Resulting linear function:

$$y = 45.8438 + 0.293123\times hours\_studied - 0.03356\times sleep\_hours + 0.19721\times attendance$$
